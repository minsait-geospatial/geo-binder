# Spatial Data Models

## Topology validation

```python
# Sí el gpkg solo tuviera una sola layer podemos usar
import geopandas as gpd
import fiona
file = "E://Datasets//GIS//CONSTRU2.gpkg"
gdf = gpd.read_file(file)
gdf["is_valid"]="Null"
gdf.head()
```

```python
import folium 

m = folium.Map([40.529, -3.641], zoom_start=19)
tile = folium.TileLayer(
        tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
        attr = 'Esri',
        name = 'Esri Satellite',
        overlay = False,
        control = True
       ).add_to(m)
folium.GeoJson(gdf).add_to(m)
m
```

```python
from shapely.geometry import Polygon
geometry = [Polygon(x) for x in gdf.geometry]
for index, item in enumerate(geometry): 
    gdf["is_valid"].iloc[index]=item.is_valid
gdf.head(-1)
```

```python
import folium 

m = folium.Map([40.529, -3.641], zoom_start=19)
tile = folium.TileLayer(
        tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
        attr = 'Esri',
        name = 'Esri Satellite',
        overlay = False,
        control = True
       ).add_to(m)

folium.GeoJson(gdf.to_json(), style_function=lambda x :{'fillColor':'green' if x['properties']['is_valid']==True else 'red', 'fillOpacity' : 0.5,}).add_to(m)
m
```
